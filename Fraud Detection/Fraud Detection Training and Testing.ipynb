```python
import pandas as pd
import numpy as np
```

```python
data_1 = pd.read_csv("creditcard-p1.csv")
data_2 = pd.read_csv("creditcard-p2.csv")
data_1.head()
```

```python
data_2.head()
```

```python
data = pd.concat([data_1, data_2], ignore_index=True)
data.head()
```

```python
data.shape
```

```python
#importing libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score
from imblearn.over_sampling import SMOTE
```

```python
X = data.drop('Class', axis=1)
y = data['Class']
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Handle imbalanced classes
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.3, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1-Score: {f1:.2f}')
```

Logistic regression is good, but need to check for other methods like decision trees

## Decision tree testing

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Initialize the Decision Tree model
decision_tree = DecisionTreeClassifier(random_state=42)

# Train the model
decision_tree.fit(X_train, y_train)

# Make predictions
y_pred_decision_tree = decision_tree.predict(X_test)
```

```python
precision_decision_tree = precision_score(y_test, y_pred_decision_tree)
recall_decision_tree = recall_score(y_test, y_pred_decision_tree)
f1_decision_tree = f1_score(y_test, y_pred_decision_tree)

# Print the evaluation metrics
print(f'Precision: {precision_decision_tree:.2f}')
print(f'Recall: {recall_decision_tree:.2f}')
print(f'F1-Score: {f1_decision_tree:.2f}')
```

Decision Tree has overfitting need ensemble techniques best switch to random forest

## Random Forest Testing

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Initialize the Random Forest model
random_forest = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the model
random_forest.fit(X_train, y_train)

# Make predictions
y_pred_random_forest = random_forest.predict(X_test)
```

```python
precision_random_forest = precision_score(y_test, y_pred_random_forest)
recall_random_forest = recall_score(y_test, y_pred_random_forest)
f1_random_forest = f1_score(y_test, y_pred_random_forest)

print(f'Precision: {precision_random_forest:.2f}')
print(f'Recall: {recall_random_forest:.2f}')
print(f'F1-Score: {f1_random_forest:.2f}')
```

This has overfitted again need to create another random forest with lesser number of trees

```python
random_forest2 = RandomForestClassifier(n_estimators=5, random_state=42,max_depth=10,  min_samples_split=5, min_samples_leaf=2)

# Train the model
random_forest2.fit(X_train, y_train)

# Make predictions
y_pred_random_forest2 = random_forest2.predict(X_test)
```

```python
precision_random_forest2 = precision_score(y_test, y_pred_random_forest2)
recall_random_forest2 = recall_score(y_test, y_pred_random_forest2)
f1_random_forest2 = f1_score(y_test, y_pred_random_forest2)

print(f'Precision: {precision_random_forest2:.2f}')
print(f'Recall: {recall_random_forest2:.2f}')
print(f'F1-Score: {f1_random_forest2:.2f}')
```

Random Forest is giving even better result, not case of overfitting and generally good recall i.e catching 97% of all the fraudaulent activities
