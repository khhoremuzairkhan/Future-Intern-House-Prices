```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import seaborn as sns
```

```python
df = pd.read_csv('House Price Data.csv')
```

```python
label_encoder = LabelEncoder()
```

```python
def preprocessing(df:pd.DataFrame) -> pd.DataFrame:
    df['complete_address'] = df['street'] + ', ' + df['city'] + ', ' + df['statezip'] + ', ' + df['country']
    df['complete_address'] = label_encoder.fit_transform(df['complete_address'])
    df.drop(['street', 'city', 'statezip', 'country', 'date'], axis=1, inplace=True)
    df = df.astype({'bedrooms': int, 'bathrooms':int, 'floors': int})
    return df
```

```python
df_preprocessed = preprocessing(df)
```

```python
X = df_preprocessed.drop('price',axis=1).values
```

```python
y = df_preprocessed['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

```python
scaler_features = MinMaxScaler()
```

```python
scaler_features.fit(X_train)
```

```python
X_train = scaler_features.transform(X_train)
```

```python
X_test = scaler_features.transform(X_test)
```

```python
y_train = y_train.reshape(-1,1)
```

```python
y_test = y_test.reshape(-1,1)
```

```python
scaler_target = MinMaxScaler()
```

```python
scaler_target.fit(y_train)
```

```python
y_scaled_train = scaler_target.transform(y_train)
```

```python
y_scaled_test = scaler_target.transform(y_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_scaled_train)
```

```python
y_pred_scaled = model.predict(X_test)
```

```python
y_pred_scaled = y_pred_scaled.reshape(-1,1)
```

```python
mse = mean_squared_error(y_scaled_test,y_pred_scaled)
```

```python
print(mse)
```

```python
y_pred = scaler_target.inverse_transform(y_pred_scaled)
```

```python
y_pred
```

```python
y_test_actual = scaler_target.inverse_transform(y_scaled_test)
```

```python
y_test_actual
```

```python
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted Y'])
```

```python
y_test_df = pd.DataFrame(y_test_actual, columns = ['Test True Y'])
```

```python
pred_df = pd.concat([y_test_df, y_pred_df], axis=1)
```

```python
pred_df
```

```python
pred_df.columns = ['True price', 'predicted price']
```

```python
sns.scatterplot(x='True price', y='predicted price', data=pred_df)
```

```python
r2 = r2_score(y_test_actual,y_pred)
```
